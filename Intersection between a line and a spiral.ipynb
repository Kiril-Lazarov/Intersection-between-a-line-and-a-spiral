# <center>Method for finding intersection points between a linear function and a spiral.<center> 
    
## <center>Abstract<center>

The problem discussed in this project arose quite unexpectedly for the author in connection with another [project](https://github.com/Kiril-Lazarov/Artificial-Intelligence/blob/main/Course%20Project/Non-simultaneous%20gravitational%20interaction.ipynb) in the field of physics—the problem of non-simultaneous gravitational interaction between two bodies. The existence of a finite possible speed for the propagation of information and interactions - the speed of light - necessitates calculating the effect of the time interval between the 'emission' of the gravitational signal - whether this 'gravitational signal' is interpreted as the curvature of spacetime or otherwise - and its 'reception' by another body. (According to the simulation conducted in this project, the main effect consists of the rotation of the orbit of the bodies in the direction of motion along this orbit). Calculating this time delay ultimately boiled down to finding the intersection point between a spiral and a linear function. The linear function represents the rectilinear and uniform motion of the receiving body, while the spiral represents the trajectory of the point from the expanding spherical wavefront, which will intersect in space with the rectilinear trajectory of the receiving body at a certain moment in time.


Finding this intersection point, however, turned out to be possible only through the use of a numerical method, as no analytical method exists. In the project itself, a numerical method was used, but the challenge of discovering an analytical method remained. This current project will present an attempt to formulate such an analytical method using elementary geometry of a right triangle and the central angle of a radius vector. The question of whether the proposed method is strictly analytical or a combination of numerical and analytical methods, however, remains open.
    
The author's current position on this matter is that the method is hybrid - it contains elements from both numerical and analytical methods. Specifically, finding the intersection point between a spiral and a line is reduced to finding the limit of convergence of a numerical sequence with two characteristic features: it can only be defined recursively, and its limit of convergence is an irrational number. The first characteristic - the recursive formula for the general term—is what creates the limitation of this method and introduces the mandatory iterative approach in calculating this limit. Finding the nth term of the sequence is impossible without finding all the preceding terms, as the input parameter for each subsequent term is the result of the previous one. This mandatory iterative process is what aligns the proposed method with numerical methods. However, what differentiates it is that, unlike numerical methods where the initial approximation value is arbitrary (i.e., freely chosen), here the initial value is derived analytically. Therefore, it is not arbitrary.
    
The second characteristic—the irrational number that serves as the limit of convergence—imposes the restriction (or advantage?) that calculating the intersection point is only possible within certain bounds. These bounds are actually imposed by the capabilities of the computing machine. For practical purposes, an accuracy of four to five decimal places is entirely sufficient.
    
Finally, from a broader perspective, there is logic in the limit being an irrational number—the spiral and the line do not depend on each other in any way; that is, they are not functions of each other. If they were, then this limit could be found purely analytically, and it would not be an irrational number.
    
    
## <center>Archimedean spiral<center>
    
The project examines the intersection between an $Archimedean$ $spiral$ and a line. The $Archimedean$ $spiral$ is a curve described by a radius $r$ that increases at a constant speed $v$ and rotates with a constant angular velocity $\theta$ around a given point $O$. The standard equation of an Archimedean spiral is:
    
$$r = a + b\theta,\quad\quad\quad\quad(1)$$
    
where $r$ is the radius of the spiral, $\theta$ is the angle of rotation, $a$ and $b$ are coefficients. Since intersection points will be sought here, the equation of the spiral will be broken down into components in the Cartesian coordinate system. Thus, for the $x$ and $y$ components, we will have the following equations respectively:
    
$$x_s =|\vec{R}|cos({\theta}),\quad\quad\quad\quad(2)$$
    
$$y_s =|\vec{R}|sin({\theta}),\quad\quad\quad\quad(3)$$
    
where $|\vec{R}|$ is the radius-vector and $\theta$ is the angle of rotation. Here, due to certain reasons, the coordinates $x$ and $y$ need to be expressed as functions of a variable. We will call this variable $t$, and it can be thought of as the quantity $time$. Thus, we write the magnitude of the radius vector $|\vec{R}|$ as:
    
$$\vec{R}_{(t)} = vt,\quad t \geq 0,\quad v>0.\quad\quad\quad\quad(4)$$
    
The other variable in the equation - the angle of rotation $\theta$ - can also be expressed in terms of the parameter $t$:
    
$$\theta_{(t)} = wt,\quad t \geq 0,\quad w \neq 0, \quad\quad\quad\quad(5)$$
    
where $w$ is the angular velocity, with its sign indicating the direction of rotation of the spiral - positive values of $w$ denote counterclockwise rotation, while negative values denote clockwise rotation. Thus, the equations for the $x$ and $y$ coordinates of the spiral take the form:
    
$$x_{s_{(t)}} =vtcos(\theta_0 + wt), \quad t \geq 0,\quad v>0 \quad w \neq 0, \quad 0 \leq \theta_{0} < 2\pi, \quad\quad\quad\quad(6)$$
    
$$y_{s_{(t)}} =vtsin(\theta_0 + wt), \quad t \geq 0,\quad v>0 \quad w \neq 0, \quad 0 \leq \theta_{0} < 2\pi. \quad\quad\quad\quad(7)$$
    
where $\theta_{0}$ is the initial angle of the spiral's unfolding.
    
Clearly, both equations do not have an analytical solution because isolating the variable $t$ on one side of the equation requires extracting it from the trigonometric function, which is not possible.
    
    
## <center>Linear function<center>
    
The linear function is a simple linear function of the form:
    
$$y_l = ax_l + b, \quad a \neq 0,\quad b \in (-\infty, \infty), \quad\quad\quad\quad(8)$$
    
where $a$ and $b$ are constants.

$a = -1,\quad$ $b = 8,\quad$ $v = 1,\quad$ $w = 1,\quad$ $\theta_0 = 0.\quad$

As can be seen from the graph, the intersection points are where the coordinates of the two independent functions coincide. That is, considering formulas (6), (7) and (8):

$$x_l = x_{s_{(t)}},\quad\quad\quad\quad(9)$$
$$y_l = y_{s_{(t)}},\quad\quad\quad\quad(10)$$

or

$$x_l =vtcos({\theta_0 + wt}),\quad\quad\quad\quad(11)$$
$$y_l =vtsin({\theta_0 + wt}).\quad\quad\quad\quad(12)$$

Evidently, there are infinitely many intersection points between the two objects regardless of their parameter values as $t$ approaches infinity. The task, therefore, is to find the exact values of the parameter $t$ for which equations (11) and (12) are satisfied.

The first step towards solving the problem is to find the principal solution for just one intersection point. Based on this solution, we will then formulate a strategy to generalize it for all intersection points.

## <center>Algorithm for finding the intersection point between a spiral and a line. Derivation of the formula for the general term of the numerical sequence.<center>
    
For explaining the algorithm, we will not represent the angle of rotation $\theta$ as a function of $t$. Thus, formulas (6) and (7) take the following form:
    
$$x_s =vtcos({\theta}),\quad\quad\quad\quad(13)$$
$$y_s =vtsin({\theta}).\quad\quad\quad\quad(14)$$   
    
We choose the following parameters:
    
$v = 2,\quad$ $w = 1,\quad$ $\theta_0 = -\pi/2.\quad$
    
We will show a portion of the spiral corresponding to a rotation of $\pi$ radians from the initial angle $-\pi/2$. We will also calculate three important points - $x_{max}$, $y_{0}$ and $x_i$. The point $x_i$ corresponds to a rotation angle of $\pi/2$ radians from the initial angle and lies on the $x$-axis. The linear function here will be represented by the vertical line $x_iA_0$.
    
Finding the intersection point consists of a precisely defined series of rotations of the spiral's radius vector. This series essentially involves repeating the same operation until the radius vector shifts to the position corresponding to the intersection point between the spiral and the vertical line $x_iA_0$.

```python
t  = np.linspace(0, np.pi, 2000)
v = 2
w = 1
theta_0 = -np.pi/2
angles = np.linspace(theta_0,theta_0 + np.pi, 2000)

# Coordinate points of the spiral
x = t*v*np.cos(angles)
y = t * v * np.sin(angles)

f = 10
y_max = max(y)
x_max = max(x)

y_axis_limit = 1.3*y_max

t_i = abs(theta_0)/w
x_i = t_i * v

ax = create_field(figsize=(f, f), x_lim=(-y_axis_limit, y_axis_limit), y_lim=(-y_axis_limit, y_axis_limit))

draw_algorithm(x_i, y_max, v, w, theta_0,ax, steps_count=2)
plt.plot(x, y,color='red')

plt.show()
```

## <center>Step 1<center>
    
We choose an arbitrary point $x_i$ lying on the $x$-axis. In the figure above, this is the point where the spiral intersects the $x$-axis. We also select a second point $y_0$ lying on the $y$-axis, which is also an intersection point of the spiral. Further down, different initial choices for these points will be simulated, and the impact of these choices on the behavior of the radius vector will be examined.
    
Since $x_i$ and $y_0$ both lies on the two axes, we can find their corresponding angles relative to the initial angle. The magnitudes of these angles in the direction of the spiral's unwinding are:
    
$$\theta_{x_i} = \pi/2,$$    
$$\theta_{y_0} = \pi.$$    
    
Then, from the formulas (6) and (7) for $x_0$ and $y_0$ we have:
    
$$x_{0_{t_{x_i}}} =v\frac{\theta_{x_i}}{w}cos(\theta_0 + \theta_{x_0}),\quad\quad\quad\quad(15)$$
    
$$y_{0_{t_{y_0}}} =v\frac{\theta_{y_0}}{w}sin(\theta_0 + \theta_{y_0}),\quad\quad\quad\quad(16)$$
    
noting that
    
$$t_{x_i} = \frac{\theta_{x_i}}{w},$$
    
$$t_{y_0} = \frac{\theta_{y_0}}{w},$$
    
$$v= 2,\quad w = 1,\quad\theta_0 = -\pi/2.$$
    
Substituting the values into the formulas for $x_i$ and $y_0$, we obtain the exact values:
    
$$x_i = \pi,$$
    
$$y_0 = 2\pi.$$    
    
    
We find the intersection point of the two lines parallel to the axes, which pass through $x_i$ and $y_0$, respectively. This is point $A_0$. The three points - $O$, $x_i$ and $A_0$ form a right triangle $O$$x_i$$A_0$ with a hypotenuse $O$$A_0$. Since the hypotenuse $O$$A_0$ and one leg $O$$x_i$ meet at the origin, the angle $\theta$ is a central angle. From here, the magnitude of the angle $\theta$ can easily be found as the $inverse$ $tangent$ of the ratio of the two legs:
    
$$\theta = arctan(\frac{y_0}{x_i}).\quad\quad\quad\quad(17)$$
    
In the specific example, the magnitude of the angle in degrees is $63.434948823$. From here, knowing the magnitude of the central angle, the value of $t$ can easily be found through the relationship between this angle $\theta$, the angular velocity $w$, and the parameter $t$ from formula (5). It only remains to add the initial angle $\theta_0$ to the angle $\theta$. Since it has a negative value of $-\pi/2$, the formula for $t$ will take the form:
    
$$t = \frac{\theta - \theta_0}{w}.\quad\quad\quad\quad(18)$$
    
Substituting (17) into (18) and using index $0$ for $t$ to denote that this is the initial term of the numerical sequence, the final formula of which will be derived below, we obtain:
    
$$t_0 = \frac{arctan(\frac{y_0}{x_i}) - \theta_0}{w}.\quad\quad\quad\quad(19)$$
    
Clearly, the hypotenuse intersects the spiral at a given point. Having already determined the exact value of $t_0$ - which we obtained using $\theta$ and $\theta_0$ -  we can easily find the corresponding radius vector of the spiral, whose angle matches the angle of the hypotenuse. The coordinates of the endpoint of the radius vector can be obtained by substituting equation (19) into equations (6) and (7):
    
$$x_{s_{(t_0)}} =v\frac{arctan(\frac{y_0}{x_i}) - \theta_0}{w}cos({w\frac{arctan(\frac{y_0}{x_i}) - \theta_0}{w} - \theta_0}),\quad\quad\quad\quad(20)$$
    
$$y_{s_{(t_0)}} =v\frac{arctan(\frac{y_0}{x_i}) - \theta_0}{w}sin({w\frac{arctan(\frac{y_0}{x_i}) - \theta_0}{w} - \theta_0}).\quad\quad\quad\quad(21)$$
    
After simplification:
    
$$x_{s_{(t_0)}} =v\frac{arctan(\frac{y_0}{x_i}) - \theta_0}{w}cos(arctan(\frac{y_0}{x_i}) - 2\theta_0),\quad\quad\quad\quad(22)$$
    
$$y_{s_{(t_0)}} =v\frac{arctan(\frac{y_0}{x_i}) - \theta_0}{w}sin(arctan(\frac{y_0}{x_i}) - 2\theta_0).\quad\quad\quad\quad(23)$$
    
Finaly, from (15) and (16) we set:
    
$$\frac{y_0}{x_i} = \frac{\theta_{y_0}sin(\theta_0 + \theta_{y_0})}{\theta_{x_i}cos(\theta_0 + \theta_{x_i})},\quad\quad\quad\quad(24)$$
    
and replace in (22) and (23):
    
$$x_{s_{(t_0)}} =v\frac{arctan(\frac{\theta_{y_0}sin(\theta_0 + \theta_{y_0})}{\theta_{x_0}cos(\theta_0 + \theta_{x_i})}) - \theta_0}{w}cos(arctan(\frac{\theta_{y_0}sin(\theta_0 + \theta_{y_0})}{\theta_{x_i}cos(\theta_0 + \theta_{x_i})}) - 2\theta_0),\quad\quad\quad\quad(25)$$
    
$$y_{s_{(t_0)}} =v\frac{arctan(\frac{\theta_{y_0}sin(\theta_0 + \theta_{y_0})}{\theta_{x_i}cos(\theta_0 + \theta_{x_i})}) - \theta_0}{w}sin(arctan(\frac{\theta_{y_0}sin(\theta_0 + \theta_{y_0})}{\theta_{x_0}cos(\theta_0 + \theta_{x_i})}) - 2\theta_0).\quad\quad\quad\quad(26)$$
    
These two long and complex formulas express the coordinates of the tip of the radius vector $\vec{R}$ as a function of time $t$. In each subsequent iteration, the corresponding values of $t$ from the numerical sequence will be used as input parameters in these formulas.
    
We conclude $step$ $1$ by determining a new value for the $y$-coordinate, $y_1$ = 4.79045... . This will be the input value for the next iteration.

Actually, the numerical sequence can be constructed for any of the parameters of the spiral - the angle $\theta$, the radius vector $\vec{R}$, or the time $t$. This is because the real variable on which all three depend is the $y$-coordinate. It would be easiest to construct the sequence directly for the varying input parameter $y$. The sequence will be denoted as {$y_n$}. The initial value of the sequence was determined in formula (16). 

$$y_{0_{t_0}} =v\frac{\theta_{y_0}}{w}sin(\theta_0 + \theta_{y_0}),\quad\quad\quad\quad(16)$$

where $\theta_0$ is the initial angle from which the spiral unfolds, and $\theta_{y_0}$ is the angle at which the spiral intersects the $y$-coordinate.

The next term will be function of the initital term:

$$y_{1_{(y_0)}} =v\frac{arctan(\frac{y_0}{x_i}) - \theta_0}{w}sin(arctan(\frac{y_0}{x_i}) - 2\theta_0).\quad\quad\quad\quad(30)$$


For the $n$-th term, we then have:


$$y_{n_{(y_{n-1})}} =v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_0}{w}sin(arctan(\frac{y_{n-1}}{x_i}) - 2\theta_0),\quad n >=1,\quad\quad\quad\quad(31)$$

or more generaly, the formula for the sequence:

$$\{y_n\} =v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_0}{w}sin(arctan(\frac{y_{n-1}}{x_i}) - 2\theta_0),\quad n >=1,\quad\quad\quad\quad(31)$$

This immediately raises the question: what is the behavior of this numerical sequence? Is it decreasing, does it have a limit, etc.? Let's start with the first aspect. We will return to the graph with the first iteration.


## <center>Monotonicity of the numerical sequence<center>

```python
ax = create_field(figsize=(f, f), x_lim=(-y_axis_limit, y_axis_limit), y_lim=(-y_axis_limit, y_axis_limit))

draw_algorithm(x_i, y_max, v, w, theta_0,ax, steps_count=2)
plt.plot(x, y,color='red')

plt.show()
```

From the graph, it can be seen that each subsequent term in the sequencе $\{y_0\}$ - $y_0$, $y_1$, $y_2$ and so on, is smaller than the previous one. The sequence can also be represented in another way — by subtracting the sine of the angle $\angle$$I$$A_0$$A_1$ from the value of the current term in the sequence.


$$y_n = y_{n-1} - IA_0sin(\pi/2 - \theta_{n-1}), \quad\quad\quad\quad(32)$$

$$IA_0 = OA_0 - \vec{R},\quad\quad\quad\quad(33)$$

$$OA_0 = \sqrt{x_i^{2} + y_{n-1}^{2}},\quad\quad\quad\quad(34)$$

$$\vec{R} = vt_{n-1},\quad\quad\quad\quad(35)$$

$$t_{n-1} = \frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_{n-1}}{w},\quad\quad\quad\quad(36)$$

$$y_n = y_{n-1} - (\sqrt{x_0^{2} + y_{n-1}^{2}} - v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_{n-1}}{w})sin(\pi/2 - \theta_{n-1}).\quad\quad\quad\quad(37)$$

To ensure that the sequence is monotonic - specifically monotonically decreasing - each successive term must be smaller than the previous one. Algebraically, this means that the second term on the right side of equation (37) - $(\sqrt{x_i^{2} + y_{n-1}^{2}} - v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_{n-1}}{w})sin(\pi/2 - \theta_{n-1})$ - must be:
1. a positive number, 
2. smaller than the preceding term $y_{n-1}$.

#### 1. The second right term is a positive number

The two factors give a positive number in two cases - when both are positive or both are negative. If we consider the second factor - $sin(\pi/2 - \theta_{n-1})$ - it will always be a positive number because, expressing $\angle$$O$$A_0$$x_i$ -the angle opposite to $\theta$ in the right triangle $O$$x_0$$A_0$, it will always be less than $\pi/2$ radians, and

$$0 < \pi/2 - \theta_{n-1}< \pi/2,\quad\quad\quad\quad(38)$$

therefore

$$sin(\pi/2 - \theta_{n-1}) > 0.\quad\quad\quad\quad(39)$$

This means that the first factor - $(\sqrt{x_i^{2} + y_{n-1}^{2}} - v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_{n-1}}{w})$ - must be a positive number too. To be positive, the following inequality must be satisfied:


$$\sqrt{x_i^{2} + y_{n-1}^{2}} > v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_{n-1}}{w}.\quad\quad\quad\quad(40)$$

Now we need to recall what the two sides of this inequality actually represent. On the graph above the point $y_0$ represent $y_{n-1}$ and $\theta_{n-1}$ - $\theta$ - from (40). The left side represents the hypotenuse $OA_0$, while the right side represents the radius vector $\vec{R}$ of the spiral, which lies on the hypotenuse. 

The point $y_0$ lies on the $y$-axis, and since it is an intersection point of the spiral, the segment $Oy_0$ is also a radius vector. Now, we will divide this radius vector into two other vectors. One radius vector will be slid to the right to the point $A_0$ – we will call this $\vec{OA_n}$, and the other will be slid along the spiral in the direction opposite to its winding – we will call this $\vec{R_n}$. Both vectors will lie on the same line.

When we slide $\vec{OA_{n-1}}$ (the vector $\vec{Oy_0}$ on the graph above) to the point $OA_n$ ($OA_n = OA_0$ on the graph), we see that its length increases.

$$\vec{OA_n} > \vec{OA_{n-1}}.\quad\quad\quad\quad(41)$$

At the same time, we rotate the radius vector $\vec{R_n}$. Since the angle of both radius vectors is smaller than the initial angle, it follows that the length of the radius vector of the spiral $\vec{R_n}$ is now smaller than the length of the initial radius vector $\vec{R_{n-1}}$, i.e., it has been rotated 'backward' along the spiral.

$$\vec{R_{n-1}} > \vec{R_n}.\quad\quad\quad\quad(42)$$

In other words, we have a situation here where two radius vectors have the same initial length but different lengths after the same operation is applied to them - one increases, and the other decreases. Formally:

$$\vec{R_{n-1}} = \vec{OA_{n-1}}.\quad\quad\quad\quad(43)$$

From (41), (42) and (43) follows that:

$$\vec{OA_n} > \vec{R_n}.\quad\quad\quad\quad(44)$$

But

$$\vec{OA_n} = \sqrt{x_i^{2} + y_{n-1}^{2}},\quad\quad\quad\quad(45)$$

$$\vec{R_n} = v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_{n-1}}{w},\quad\quad\quad\quad(46)$$

and

$$\sqrt{x_i^{2} + y_{n-1}^{2}} > v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_{n-1}}{w},\quad\quad\quad\quad(40)$$

which was to be proven.

In conclusion, both factors are positive numbers for every $y_{n-1}.$


#### 2. Currenth term $y_n$ is smaller than the preceding term $y_{n-1}$.

We need to show that:

$$y_{n-1} > y_n.\quad\quad\quad\quad(47)$$


Let's return to the graph above. Again, $y_0$ is equal to $y_{n-1}$ from (46). The geometric meaning of this statement is that each subsequent term in the sequence will be located 'further down' the vertical line, 'progressing' towards the intersection point with the spiral.

Since the two vectors - $\vec{R_n}$ and $\vec{OA_n}$ - lie on the same line and $0 < \theta_n < \pi/2$, follows that both coordinates of the endpoint of $\vec{A_n}$ will be greater than the coordinates of the endpoint $\vec{R_n}$.

$$x_{A_n} > x_{R_n},\quad\quad\quad\quad(48)$$

$$y_{A_n} > y_{R_n}.\quad\quad\quad\quad(49)$$

By setting

$$y_{n-1} = y_{A_n},\quad\quad\quad\quad(50)$$

$$y_n = y_{R_n},\quad\quad\quad\quad(51)$$

we obtain

$$y_{n-1} > y_n.\quad\quad\quad\quad(47)$$

In conclusion, the numerical sequence 

$$\{y_n\} =v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_0}{w}sin(arctan(\frac{y_{n-1}}{x_i}) - 2\theta_0),\quad n >=1,\quad\quad\quad\quad(31)$$

is monotonically decreasing. Now we need to show that it is bounded below.

## <center>Bounded below<center>
    
We need to show that
    
$$\lim_{n \to \infty} y_n = L, \quad n >=1.\quad\quad\quad\quad(52)$$
    
We have already established that the sequence is decreasing. To show that it is also bounded below means proving that the difference between two consecutive terms is allways positive.
    
$$y_{n-1} - y_n > 0.\quad\quad\quad\quad(53)$$
    
Formula (53) directly follows from (47), but we will take a different approach. Since each iteration divides the initial radius vector into two radius vectors, as we have shown, the difference between the two vectors will always be a positive number. This difference is the hypotenuse in the small triangle formed at each iteration between the two coordinate axes and the line segment on which the two vectors lie. This is the triangle $IA_0A_1$ on the graph above. The difference $y_{n-1} - y_n$ ($y_0 - y_1$ on the graph) is actually the adjacent leg to this hypotenuse. Using index $n$ for the general case we can express this hypotenuse as the difference between the two vectors.
    
$$I_nA_n = \vec{OA_n} - \vec{R_n}.\quad\quad\quad\quad(54)$$
    
Therefore
    
$$y_{n-1} - y_n = (\vec{OA_n} - \vec{R_n})sin(\pi/2 - \theta_n).\quad\quad\quad\quad(55)$$
    
    
Since 
    
$$(\vec{OA_n} - \vec{R_n}) > 0,\quad\quad\quad\quad(56)$$
    
$$sin(\pi/2 - \theta_n) > 0,\quad\quad\quad\quad(57)$$
    
then
    
$$y_{n-1} - y_n > 0.\quad\quad\quad\quad(53)$$
    
which was to be proven.
    
Thus, the numerical sequence is monotonically decreasing and bounded below. Its limit is an irrational number.
    
Finally, we need to prove that the decreasing sequence will not overshoot the intersection point between the line and the spiral. Overshooting the intersection point would mean that the limit of the sequence ends up on the other side of the line, or in other words, the $x$-coordinate of the spiral would be greater than the $x$-coordinate of the line.
    
$$I_{x_n} < x_0,\quad\quad\quad\quad(58)$$
    
where $I_{x_n}$ is the $x$-coordinate of the point $I_n$ which is the head of $\vec{R_n}$. 
    
From formulas (56) and (57), which state that the hypotenuse $I_nA_{n-1}$ and the leg $A_{n-1}A_n$ of the small triangle $I_nA_{n-1}A_n$ exist, it follows that the other leg $I_nA_n$ also exists. Since the angle $\theta_n$ decreases with each iteration, the opposite angle $\pi/2 - \theta_n$ must increase. Accounting for the fact that
    
$$I_nA_n = I_nA_{n-1}cos(\pi/2 - \theta_n),\quad\quad\quad\quad(59)$$
    
$$A_{x_n} = x_0,\quad\quad\quad\quad(60)$$
    
$$I_nA_n = x_0 - I_{x_n},\quad\quad\quad\quad(61)$$
    
follows that
    
$$x_0 - I_{x_n} =  I_nA_{n-1}cos(\pi/2 - \theta_n),\quad\quad\quad\quad(62)$$
    
$$I_nA_{n-1} > 0,\quad\quad\quad\quad(63)$$
    
$$cos(\pi/2 - \theta_n) > 0,\quad\quad\quad\quad(64)$$
    
$$x_0 - I_{x_n} > 0,\quad\quad\quad\quad(65)$$
    
so
    
$$I_{x_n} < x_0,\quad\quad\quad\quad(66)$$
    
which was to be proven.

Theoretically, this point could be found by using the slope of the spiral at that point, i.e., its derivative. However, finding this derivative encounters the same problem addressed by this presentation - it does not have an analytical solution. Let's check this for the $y$ coordinate of the spiral.

$$y_{s_{(t)}} =vtsin({wt - \theta_0}), \quad t \geq 0,\quad v>0, \quad w \neq 0, \quad 0 \leq \theta_{0} < 2\pi. \quad\quad\quad\quad(7)$$

Regardless of what value for the derivative $y'_{s_{(t)}}$ we are seeking - for example, $y'_{s_{(t)}} = 0$ - this equation does not have an analytical solution for the same reason that its antiderivative does not have an analytical solution: we cannot isolate the variable $t$ on one side of the equation. What remains to be done is to find this value using brute force and arbitrary precision.

This arbitrary precision depends on two factors - one is the parameter $num$ of the $.linspace()$ function, as we have already mentioned, and the other is the number of iterations $steps\_count$ of the numerical sequence. Both factors are directly proportional to the precision: the larger they are, the greater the precision.

To speed up the execution of the algorithm, only the final values of the intersection point and its corresponding radius vector will be shown on the graph.

```python
num = 10000
t_c  = np.linspace(0, np.pi, num)
angles = np.linspace(theta_0,theta_0 + np.pi, num)
x_p = max(t_c*v*np.cos(angles))
print('x_p max: ', x_p, 'x_max: ', x_max)
ax = create_field(figsize=(f, f), x_lim=(-y_axis_limit, y_axis_limit), y_lim=(-y_axis_limit, y_axis_limit))

draw_algorithm(x_p, y_max, v, w, theta_0,ax, 
               only_result=True,
               steps_count=35000)
plt.plot(x, y,color='red')
plt.show()
```

### <center>Investigation of the Behavior of the Sequence Limit<center>

#### <center>Alternative Definition of a Spiral<center>
    
The described method for finding the intersection point between a line and a spiral allows for the formulation of yet another definition of a spiral curve through the limit of a numerical sequence. The formulation used so far represented the spiral curve as a function of the parameter $t$.
    
$$x_{s_{(t)}} =vtcos({wt - \theta_0}), \quad t \geq 0,\quad v>0, \quad w \neq 0, \quad 0 \leq \theta_{0} < 2\pi, \quad\quad\quad\quad(6)$$
    
$$y_{s_{(t)}}=vtsin({wt - \theta_0}), \quad t \geq 0,\quad v>0, \quad w \neq 0, \quad 0 \leq \theta_{0} < 2\pi. \quad\quad\quad\quad(7)$$
    
The vertical line, whose intersection point with the spiral we found as the limit of a numerical sequence, was given as a single value along the $x$-axis. This allows every point on the spiral curve to be found as a function of two parameters: $x$ and $y$. Now we will investigate the behavior of $x$ and $y$ by keeping one parameter constant and varying the other. The method of constructing the numerical sequence actually allows us to find direct relationships between them, avoiding the necessity of knowing the value of the parameter $t$.

The reason it is possible to avoid the parameter $t$ is that the arbitrary choice of $x$ and $y$ also determines a precise initial angle $\theta$ between them. Determining the initial angle, in turn, allows for the exact value of an initial parameter $t_0$ to be specified through the following relationship:
    
$$t_0 = \frac{\theta_0}{w}. \quad\quad\quad\quad(68)$$
    
Determining the initial value of $t_0$ naturally leads to the determination of the subsequent value of $x$ or $y$ (in the explanations above, we assumed $x$ as a constant and $y$ as a variable). In turn, this leads to the next value of $t_1$, and so on. Therefore, we can write more general
    
$$t = f(x, y).\quad\quad\quad\quad(69)$$
    
In a certain sense, this formula serves as the inverse function of (6) and (7) because, as we have seen, all the points defined by the numerical sequence lie on the spiral curve.
    
If we denote a given point on the spiral as $S$, then it can be represented as a function of $x$ and $y$
    
$$S_{(x, y_n)} = v\frac{arctan(\frac{y_{n-1}}{x}) - \theta_0}{w}sin(arctan(\frac{y_{n-1}}{x}) - 2\theta_0),\quad x=const, \quad n >=1,\quad\quad\quad\quad(70)$$ 

$$S_{(y, x_n)} = v\frac{arctan(\frac{x_{n-1}}{y}) - \theta_0}{w}sin(arctan(\frac{x_{n-1}}{y}) - 2\theta_0),\quad y=const, \quad n >=1,\quad\quad\quad\quad(71)$$ 
    
while keeping in mind that the expression on the right - $v\frac{arctan(\frac{y_{n-1}}{x}) - \theta_0}{w}sin(arctan(\frac{y_{n-1}}{x}) - 2\theta_0)$ - is the formula for the n-th term of the sequence from (31)
    
$$\{y_n\} =v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_0}{w}sin(arctan(\frac{y_{n-1}}{x_i}) - 2\theta_0),\quad n >=1,\quad\quad\quad\quad(31)$$
    
Now we will examine graphically the behavior of $S$ for different values of $x$ and $y$ as constants and variables.

```python
ax = create_field(figsize=(8, 8), x_lim=(-y_axis_limit, y_axis_limit), y_lim=(-y_axis_limit, y_axis_limit))

draw_algorithm(x_max-1, y_max, v, w, theta_0,ax,only_result=True, steps_count=100, x_text_coef=5)
plt.plot(x, y,color='red')

plt.show()
```

Here we see that for an input value of $x$ less than the point of maximum displacement, the limit of the sequence $S$ is the point with coordinates (2.64, 4.507953332), i.e., the $x$ coordinate of $S$ is the input value itself. As we saw earlier, this also applies to the case when the input value is equal to the maximum displacement. Now, let's see what happens to $S$ when we choose $x > x_{max}.$

```python
ax = create_field(figsize=(8, 8), x_lim=(-y_axis_limit, y_axis_limit), y_lim=(-y_axis_limit, y_axis_limit))

draw_algorithm(x_max+1, y_max, v, w, theta_0,ax,only_result=True, steps_count=100, x_text_coef=3)
plt.plot(x, y,color='red')

plt.show()
```

### <center>Finding the N-th Intersection Point<center>
    
Determining which coil of the spiral the limit of the numerical sequence will 'land on' depends on the magnitude of the chosen input value for $y$, i.e., on the magnitude of the spiral's radius vector. This, in turn, is a function of $t$.
    
$$\vec{R}_{(t)} = vt,\quad t \geq 0,\quad v>0.\quad\quad\quad\quad(4)$$
    
Before we proceed to derive a formula for the n-th intersection point, we need to address another issue with the algorithm. The graph below illustrates a case where the numerical sequence does not converge to a limit.

```python
v_line_x = -8 # the x-coordinate of the vertical line
v = 7
w = 1
theta_0 = 0.1*np.pi/2
show_all_intersects(v,w, theta_0, v_line_x,
                    iterations_count=5,
                    nth_intersect=1,
                    correction_mech=False,
                    show_on_screen=True)
```

The graph shows the first six iterations of the algorithm. It is evident that the spiral points (the purple dots on the graph) with each successive iteration moves "forward" along the spiral curve, rather than "backward" toward its intersection with the vertical line. With the spiral and vertical line parameters set as they are, the algorithm calculates each new $x$-coordinate to be greater by absolute value than the $x_{line}$-coordinate where the line is defined, meaning the corresponding point on the spiral is "outside" the rectangle defined by $x_{line}$ and $y$.

Something interesting happens at the next, seventh iteration. The last radius vector - the one whose tip ends at the blue dot on the graph - moves backward after landing in the third quadrant during the current sixth iteration.

```python
v_line_x = -8 # the x-coordinate of the vertical line
v = 7
w = 1
theta_0 = 0.1*np.pi/2
show_all_intersects(v,w, theta_0, v_line_x,
                    iterations_count=6,
                    nth_intersect=1,
                    correction_mech=False,
                    show_on_screen=True)
```

We will provide an explanation for this shortly, but the reader is encouraged to independently explore the behavior of the blue dot by adjusting the $iterations\_count$ parameter in the $show\_all\_intersects()$ function. What the reader will observe is that the blue point begins to oscillate around the negative $x$-axis. If the parameters of the function $show\_all\_intersects()$ are set as follows: $iterations\_count=None$ and $show\_on\_screen=True$, the blue dot will 'land' on the negative $x$-axis.

The reason the algorithm 'fails' here is that it cannot maintain the constant value of the new $x$-coordinate of the radius vector, ensuring it remains smaller than the $x$-coordinate of the vertical line. As we pointed out, in this particular case, the algorithm returns a greater absolute value of $x_{spiral vector}$ with each subsequent iteration. The fix for this 'bug' consists of a simple check at each iteration to see if the new $x_{spiral vector}$ is greater in absolute value than the $x_{line}$. If it is, then the value of $x_{spiral vector}$ is reset to the value of $x_{line}$. 

This behavior of the algorithm necessitates restricting the domain of the proof for the monotonicity and boundedness of the numerical sequence. The numerical sequence {$y_n$} is monotonically decreasing and bounded (either from below or above) if the absolute value of the point on the vertical line $x_l$ is greater than a given value $a$. What this value is and how to proceed in cases when this condition is not met will be discussed later in the presentation, along with another problem that arises under these conditions. For now, the correct functioning of the algorithm will rely on a purely programming solution: checking the value of $x_l$ at each iteration and correcting it if necessary, as indicated in the paragraph above.

Now we can proceed to the issue of finding the n-th intersection point. The first challenge in this task is to assign a "label" to each of them.


#### <center>Arrangement of intersection points<center>
    
    
Since the radius vector of the spiral is a function of time $t$, the intersection points with the line should also be arranged in sequence relative to $t$. Thus, each intersection point naturally has its unique value relative to $t$. The first step in finding the n-th intersection point involves locating the intersection points of the spiral with the $y$-axis, which is essentially also a line.

```python
v_line_x = -8
v = 1
w = 1
theta_0 = 3*np.pi/2
show_all_intersects(v,w, theta_0, v_line_x,
                    show_line_and_itrscts=False,
                    show_labels=True)
```

It can be seen that the intersection points on the $y$-axis are sequentially arranged every 180 degrees. This allows for the construction of a numerical sequence where the n-th term is the intersection point of the spiral with the $y$-axis. Since the points are a functions of the parameter $t$, the numerical sequence will be constructed based on the parameter $t$.

$$\{T_n\} = t_0 + (n-1)\frac{\pi}{|w|},\quad w\neq0, \quad n>0, \quad\quad\quad\quad(72)$$

where $t_0$ is the the 'time' of the first intersection point of the spiral with the ordinate axis, and $\frac{\pi}{|w|}$ represents the amount of time with respect to the angular velocity $w$ that needs to be added to $t_0$ to find the n-th intersection point time. Here, we take the absolute value of $w$ because the time $t$ is defined as a positive quantity. So, when $n = 1$, we have only the initial moment $t_0$, when the spiral first intersects the $y$-axis. With each subsequent $n$, we simply add $n-1$ times the time it takes for the radius vector of the spiral to rotate by $\pi$ radians.

#### <center>Finding the first intersection point of the spiral with the ordinate axis<center>
    
The time of the first intersection point $t_0$ depends on two parameters: the initial angle $\theta_0$ and the angular velocity $w$ - both its magnitude and direction. As mentioned earlier, the positive and negative values of the angular velocity $w$ determine the direction of rotation of the spiral - counterclockwise and clockwise, respectively. For convenience and simplicity in calculations, the initial angle will take only positive values between 0 and $2\pi$, i.e.
    
$$0<=\theta_0<2\pi.\quad\quad\quad\quad(73)$$
    
Since we are looking for the intersection point with the $y$-axis, we will set the initial angle values to be multiples of $\pi/2$
    
$$\theta_0=k\frac{\pi}{2}, \quad0<=k<4.\quad\quad\quad\quad(74)$$
    
Therefore, there are only two possible options for the first intersection point - it either lies on the positive $y$-axis, i.e. angle of rotation $\pi/2$, or on the negative $y$-axis, i.e. $3\pi/2$. The task here is to determine which of the two cases corresponds to the value of the initial angle $\theta_0$ and the direction of the spiral's rotation - the sign in front of $w$.
    
The first step is to determine which coordinate axis comes next in the direction of the spiral's winding, based on the value of the initial angle and the direction in which the spiral is expanding.

```python
show_first_intersect_axis(3.5)
```

On the graph above, we see that with an initial angle of 315 degrees - indicated by the black radius vector; and a positive angular velocity $w$ - meaning the spiral unwinds in a counterclockwise direction, the next coordinate axis is the positive $x$-axis.

The formula for finding the difference between the initial angle and the next coordinate axis is:

$$\Delta \theta = \lfloor k + 1 \rfloor\frac{\pi}{2} - \theta_0. \quad\quad\quad\quad(75)$$

Replacing $\theta_0$ from (74)

$$\Delta \theta_{(k)}  =\lfloor k + 1 \rfloor\frac{\pi}{2}- k\frac{\pi}{2}, \quad\quad\quad\quad(76)$$

$$\Delta \theta_{(k)}  = \frac{\pi}{2}(\lfloor k + 1 \rfloor- k). \quad\quad\quad\quad(77)$$

The formula indicates the magnitude of the angle by which the radius vector needs to rotate, starting from the initial angle, in order to reach the next coordinate axis. Since we only need the ordinate axis, an additional term equal to $\pi/2$ must be introduced into the formula and added to the rotation angle $\Delta\theta$.

$$\Delta \theta_{(k)} = \frac{\pi}{2}(\lfloor k + 1 \rfloor- k) + D\frac{\pi}{2}, \quad\quad\quad\quad(78)$$

$$\Delta \theta_{(k)} = \frac{\pi}{2}(\lfloor k + 1 \rfloor- k + D). \quad\quad\quad\quad(79)$$

The coefficient $D$ determines whether the additional angle $\pi/2$ should be added or not. Therefore, it will only have two values - 0 and 1. Whether $D$ will be 0 or 1 depends on the coefficient $k$ that defines the initial angle as a percentage of $\pi/2$.

$$D = f(k). \quad\quad\quad\quad(80)$$

However, the coefficient $k$ is a continuous variable in the interval $[0,4)$. The task is to reduce it to 0 or 1. In the graph below, it has a value of 3.5, and since we are considering a positive angular velocity, the radius vector will cross the positive $y$-axis.

```python
k = 3.5
show_first_intersect_axis(k,D_coeffs=True)
```

The reader can try setting their own values for $k$ within the specified interval. It is evident that there is a need to add an angle $\pi/2$ arises when the integer part of $k$ is odd. We can call this function a $binary function$ and denote it as $D^{[0,1]}_{(k)}$. The ordered pair $[0, 1]$ in the index specifies the two possible outcomes, where the first position shows the result when the integer part of $k$ is an even number, and the second position shows the result when integer part of $k$ is an odd number. Therefore

$$D^{[0,1]}_{(k)} = \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2}. \quad\quad\quad\quad(81)$$

$$D^{[0,1]}_{(k)} = 0, \quad\frac{\lfloor k \rfloor}{2} \in \mathbb{Z}, \quad\quad\quad\quad(82)$$

$$D^{[0,1]}_{(k)} = 1, \quad\frac{\lfloor k \rfloor}{2} \notin \mathbb{Z}. \quad\quad\quad\quad(83)$$

And so, from here for $\Delta\theta$ we obtain

$$\Delta \theta_{(k)} = \frac{\pi}{2}(\lfloor k + 1 \rfloor- k + D^{[0,1]}_{(k)}). \quad\quad\quad\quad(84)$$

$$\Delta \theta^+_{(k)} = \frac{\pi}{2}(\lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2}). \quad\quad\quad\quad(85)$$

We add the symbol $+$ to the index of $\Delta \theta_{(k)}$ to indicate that the direction of the spiral is positive - $w > 0$.

As seen from the graph, in the opposite case - when the spiral rotates clockwise - the rotation angle should be the complementary angle to 180 degrees relative to $\Delta \theta^+$. In other words, we will express the difference $\Delta \theta^-$ as a function of $\Delta \theta^+$.

$$\Delta \theta^- = f(\Delta \theta^+_{(k)}), \quad\quad\quad\quad(86)$$

$$f(\Delta \theta^+_{(k)}) = c\pi \pm \Delta \theta^+_{(k)}.\quad\quad\quad\quad(87)$$

The coefficient $c$ takes on two values - 0 or 1, and which of the two signs is determined by the same coefficient. The table below shows the values of $\Delta \theta^-$ relative to $\Delta \theta^+$.

```python
show_init_theta_dependencies(show_expr=False)
```

As can be seen, for integer values of $k$, $\Delta \theta^+$ and $\Delta \theta^-$ are equal in magnitude, whereas for fractional values $\Delta \theta^-$ = 180 - $\Delta \theta^+$. We remind that the values of $\Delta \theta^+$ and $\Delta \theta^-$ represent the respective angular distances to the nearest ordinate axis in the direction of the spiral's rotation.

The task now is to construct a formula that accurately calculates $\Delta \theta^-$. In the table below, we can see how the signs and the coefficient $c$ should appear.

```python
show_init_theta_dependencies(show_expr=True)
```

In the fourth column we have an alternation of 0 and 1 before $\pi$, and a corresponding alternation of $+$ and $-$. Therefore, the task now is to derive a function that takes a real number $k$ and returns 0 or 1 depending on whether $k$ is an integer or a fraction. 
Before we do that, let's note that the coefficient $D^{[0, 1]}_{(k)}$ we derived earlier won't be useful here because it returns two alternating zeros and ones.

```python
show_d_coeffs()
```

So, let’s define a function that converts integer values of $k$ to 0 and fractional values to 1. Here we can also use a $binary function$, which we will denote as $B^{[0,1]}_{(k)}$.

$$B^{[0,1]}_{(k)} = \lceil k \rceil - \lfloor k \rfloor,\quad k \in \mathbb{R},\quad\quad\quad\quad(88)$$

$$B^{[0,1]}_{(k)} = 0,\quad k \in \mathbb{Z},\quad\quad\quad\quad(89)$$

$$B^{[0,1]}_{(k)} = 1,\quad k \notin \mathbb{Z}.\quad\quad\quad\quad(90)$$

We will return to this formula after the main exposition to examine its behavior in different situations and to demonstrate how it can be used to generate the sequence of prime numbers.

Armed with the $binary function$, we return to the formula (87) where we substitute $c$ with $B^{[0,1]}_{(k)}$.

$$\Delta \theta^- = B^{[0,1]}_{(k)}\pi \pm \Delta \theta^+_{(k)}.\quad\quad\quad\quad(91)$$

We need to handle the signs. To determine which of the two signs to use in specific situations, we can once again use the $binary function$. In this case, the function should return values of 1 or -1:

$$S^{[1, -1]}_{(k)} = 1 - 2B^{[0,1]}_{(k)},\quad k \in \mathbb{R},\quad\quad\quad\quad(92)$$

$$S^{[1, -1]}_{(k)} = 1 - 2(\lceil k \rceil - \lfloor k \rfloor),\quad\quad\quad\quad(93)$$

$$S^{[1, -1]}_{(k)} = 1,\quad k \in \mathbb{Z},\quad\quad\quad\quad(94)$$

$$S^{[1, -1]}_{(k)} = -1,\quad k \notin \mathbb{Z}.\quad\quad\quad\quad(95)$$

In this case, the formula (87) takes the form:

$$\Delta \theta^- = B^{[0,1]}_{(k)}\pi +S^{[1, -1]}_{(k)} \Delta \theta^+,\quad\quad\quad\quad(96)$$

or in terms only of $k$:

$$\Delta \theta^-_{(k)} = (\lceil k \rceil - \lfloor k \rfloor)\pi +(1 - 2(\lceil k \rceil - \lfloor k \rfloor))\Delta \theta^+_{(k)}.\quad\quad\quad\quad(97)$$

Now we have two separate formulas for the difference between the initial angle and the next ordinate axis in the direction of the spiral's rotation - $\Delta \theta^+_{(k)}$ and $\Delta \theta^-_{(k)}$. The goal is to combine these two formulas into a single one. 

To do this, let's return to the table of angles.

```python
show_init_theta_dependencies(show_expr=True, include_b=True)
```

This time, we see the entire formula for $\Delta \theta^-_{(k)}$ with the inclusion of both the binary and sign coefficients. To incorporate the effect of the spiral's direction of development, we need to introduce two numbers that represent its numerical orientation. These numbers are again 1 and -1, which are derived as a function of the angular velocity $w$.

$$d = \frac{w}{|w|},\quad w \neq 0,\quad\quad\quad\quad(98)$$

$$d = 1,\quad w>0,   \quad\quad\quad\quad(99)$$

$$d = -1,\quad w<0.   \quad\quad\quad\quad(100)$$

Before we introduce the angular velocity into the formula, we need to make a modification. The sign coefficient $S^{[1, -1]}_{(k)}$ in front of  $\Delta \theta^+$ from (96) should be removed. This can be achieved by performing some algebraic transformations.

$$\Delta \theta^- = B^{[0,1]}_{(k)}\pi +S^{[1, -1]}_{(k)} \Delta \theta^+.\quad\quad\quad\quad(96)$$

We will substitute $S^{[1, -1]}_{(k)} $ from equation (93) into equation (96) and simplify the expression.

$$S^{[1, -1]}_{(k)} = 1 - 2(\lceil k \rceil - \lfloor k \rfloor),\quad\quad\quad\quad(93)$$

$$\Delta \theta^- = B^{[0,1]}_{(k)}\pi +(1 - 2(\lceil k \rceil - \lfloor k \rfloor))\Delta \theta^+,\quad\quad\quad\quad(101)$$

$$\Delta \theta^- = B^{[0,1]}_{(k)}\pi +\Delta \theta^+ - 2\Delta \theta^+(\lceil k \rceil - \lfloor k \rfloor).\quad\quad\quad\quad(102)$$

Now we have $\Delta \theta^+$ as a term without a coefficient in front of it. The last term has the binary function as its coefficient after $- 2\Delta \theta^+$. We can factor out this $binary function$.

$$\Delta \theta^- = B^{[0,1]}_{(k)}\pi +\Delta \theta^+ - 2\Delta \theta^+B^{[0,1]}_{(k)},\quad\quad\quad\quad(103)$$

$$\Delta \theta^- = B^{[0,1]}_{(k)}(\pi- 2\Delta \theta^+) +\Delta \theta^+.\quad\quad\quad\quad(104)$$

Obviously, if:

$$B^{[0,1]}_{(k)}(\pi- 2\Delta \theta^+) = 0,\quad\quad\quad\quad(105)$$

then

$$\Delta \theta^- = \Delta \theta^+.\quad\quad\quad\quad(106)$$

In other words, what we need here is a function that returns 0 for a positive angular velocity and 1 for a negative angular velocity. We will now define another $binary function$ $W^{[0, 1]}_{(w)}$ that takes the angular velocity as its argument.

$$W^{[0, 1]}_{(w)} = \frac{1 - d}{2},\quad\quad\quad\quad(107)$$

$$W^{[0, 1]}_{(w)} = \frac{1 - \frac{w}{|w|}}{2},\quad w \neq 0,\quad\quad\quad\quad(108)$$

$$W^{[0, 1]}_{(w)} = 0,\quad w>0,\quad\quad\quad\quad(109)$$

$$W^{[0, 1]}_{(w)} = 1,\quad w<0.\quad\quad\quad\quad(110)$$

Let add (108) into (104)

$$\Delta \theta_{(w,k)} = W^{[0, 1]}_{(w)}B^{[0,1]}_{(k)}(\pi- 2\Delta \theta^+) +\Delta \theta^+.\quad\quad\quad\quad(111)$$

We removed the negative index of $\Delta \theta^-$ and expressed the general solution for $\Delta \theta$ as a function of $w$ and $k$.

The complete formula - only in terms of $w$ and $k$ - will then have the following form:

$$\Delta \theta_{(w,k)} = \frac{1 - \frac{w}{|w|}}{2}(\lceil k \rceil - \lfloor k \rfloor)(\pi- 2\frac{\pi}{2}(\lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2})) +\frac{\pi}{2}(\lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2}),\quad\quad\quad\quad (112)$$

$$\Delta \theta_{(w,k)} = \frac{1 - \frac{w}{|w|}}{2}(\lceil k \rceil - \lfloor k \rfloor)(\pi- \pi(\lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2})) +\frac{\pi}{2}(\lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2}),\quad\quad\quad\quad (113)$$

$$\Delta \theta_{(w,k)} = \frac{1 - \frac{w}{|w|}}{2}(\lceil k \rceil - \lfloor k \rfloor)(\pi(1 - \lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2})) +\frac{\pi}{2}(\lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2}).\quad\quad\quad\quad (114)$$

The full expression of this formula is quite lengthy, so from now on, we will use the shorter version from (111).

The two binary functions - $B^{[0,1]}_{(k)}$ and $W^{[0, 1]}_{(w)}$ - act as 'switches.' Depending on the argument passed to them, they either cancel or retain the term $(\pi- 2\Delta \theta^+)$.

$$\{T_n\} = t_0 + (n-1)\frac{\pi}{|w|},\quad w\neq0, \quad n>0. \quad\quad\quad\quad(72)$$

Knowing the angular difference $\Delta \theta_{(w, k)}$, we can easily find the time $t_0$ corresponding to it based on the magnitude of the angular velocity $|w|$.

$$t_0 = \frac{\Delta \theta_{(w, k)}}{|w|},\quad w\neq0, \quad n>0. \quad\quad\quad\quad(115)$$

We substitute into (72)

$$\{T_n\} = \frac{\Delta \theta_{(w, k)}}{|w|} + (n-1)\frac{\pi}{|w|},\quad w\neq0, \quad n>0, \quad\quad\quad\quad(116)$$

$$\{T_n\} = \frac{\Delta \theta_{(w, k)} + (n-1)\pi}{|w|},\quad w\neq0, \quad n>0. \quad\quad\quad\quad(117)$$

This is the final formula for the n-th term of the numerical sequence representing the times at which the spiral intersects the ordinate axis. Each term in this sequence serves as the initial value that needs to be input into the algorithm for finding the intersection of the spiral and the linear function. It is important to note that the algorithm itself determines another numerical sequence.

$$\{y_n\} =v\frac{arctan(\frac{y_{n-1}}{x_i}) - \theta_0}{w}sin(arctan(\frac{y_{n-1}}{x_i}) - 2\theta_0),\quad n >=1.\quad\quad\quad\quad(31)$$

This means that it is necessary to construct a numerical sequence composed of other numerical sequences, or in short - a two-dimensional numerical sequence. It can be viewed as a table, with the initial values along the first row and the points of the spiral that increasingly approach the line along the columns.

$T_n$ is the sequential number in the numerical sequence of the times $t$ at which the spiral intersects the ordinate axis and serves as the input parameter for the numerical sequence {$t_{nm}$} of the algorithm to find the boundary of the radius vector of the spiral as it approaches the line.

The next task now is to modify the formula for the general term of the sequence {$t_{nm}$} with respect to the input parameter $T_n$ provided by {$T_n$}. The current formula (31), which we derived for {$y_n$}, was derived solely for the purpose of explaining the algorithm.

First, we need to change the way we calculate the angle between the radius vector of the spiral and the linear function. The term $arctan(\frac{y_{n-1}}{x_i})$ has two limitations. The first is that

$$arctan(\frac{y_{n-1}}{x_i}) = undefined,\quad x_i = 0.  \quad\quad\quad\quad(118)$$

The second is that $arctan$ is limited to angles from 0 to 90 degrees, which is not applicable for the final general solution.



### <center>Step 1<center>

In the above graph, $\theta$ is the angle we used to measure the angular displacement of the radius vector to its intersection point with the linear function. Now, however, we will measure the angle $\Delta \phi$. Its value will be calculated using the law of cosines, as we know the lengths of all three sides of the triangle $\Delta OA_0y_0$.  $\Delta \phi$ is the angle by which we need to rotate the initial radius vector $\vec {OA_0}$ - i.e., the one provided as the starting value for initiating the algorithm through the time $T_n$ from the numerical sequence {$T_n$} - to take the next step toward bringing it closer to the intersection point with the line. Start with the law of cosines.
    
$$c^2 = a^2 + b^2 - 2abcos\Delta \phi, \quad\quad\quad\quad(119)$$
    
where $c$ is the leg $A_0y_0$, $a$ is the hipotenuse $OA_0$, and $b$ is the other leg $y_0O$.
    
$$(A_0y_0)^2 = (OA_0)^2 + (y_0O)^2 - 2(OA_0)(y_0O)cos\Delta \phi, \quad\quad\quad\quad(120)$$
    
$$cos\Delta \phi = \frac{(A_0y_0)^2 - (OA_0)^2 - (y_0O)^2}{ - 2(OA_0)(y_0O)}, \quad\quad\quad\quad(121)$$
    
$$cos\Delta \phi = \frac{ (OA_0)^2 + (y_0O)^2 - (A_0y_0)^2}{2(OA_0)(y_0O)}. \quad\quad\quad\quad(122)$$
    
We begin by sequentially substituting the triangle's side lengths with the actual values of the line and the spiral.
    
The side $y_0O$ is the initial value of the radius vector $\vec{R_{(t_n)}}$, which is a function of $t_n$.

$$y_0O = \vec{R_{(t_n)}},\quad n \geq 1, \quad\quad\quad\quad (123)$$

$$\vec{R_{(t_n)}} = vt_n,\quad n \geq 1. \quad\quad\quad\quad (124)$$

$A_0y_0$ is the horizontal leg, which is the difference between the point where the vertical line is positioned and the current $x$-component of the radius vector $\vec{R_{(t_n)}}$.

$$A_0y_0 = |x_l - x_{(t_n)}|,\quad\quad\quad\quad (125)$$


$OA_0$ is the hypotenuse of the triangle $\Delta OA_0y_0$, whose length is:

$$OA_0 = \sqrt {y_{(t_n)}^2 + x_l^2},\quad\quad\quad\quad (126)$$

$$y_{(t_n)} = vt_nsin(\theta_0 + wt_n).\quad\quad\quad\quad (127)$$

In this case, $y_{(t_n)}$ coincides with the length of $\vec{R_{(t_n)}}$ because it is the initial value of the radius vector, which lies on the ordinate axis. In the subsequent iterations of the algorithm, $y_{(t_n)}$ changes.

So, we substitute the actual values in (122).

$$cos\Delta \phi = \frac{ (\sqrt {y_{(t_n)}^2 + x_l^2})^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2(\sqrt {y_{(t_n)}^2 + x_l^2})(vt_n)}. \quad\quad\quad\quad(128)$$

And since we are looking for the angle $\Delta \phi_{(t_n)}$ itself:

$$\Delta \phi_{(t_n)} = arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}}), \quad y_{(t_n)}^2 + x_l^2 \neq 0, \quad\quad\quad\quad(129)$$
    

where $x_{(t_n)}$ is the $x$-component of the radius vector $\vec{R_{(t_n)}}$
    
$$x_{(t_n)} = vt_ncos(\theta_0 + wt_n).\quad\quad\quad\quad(130)$$

    
We will leave formula (129) as it is for the sake of clarity, even though it could be simplified to some extent.
    
    
Thus, we arrive at step 2 - determining the sign of $\Delta \phi_{(t_n)}$, i.e., whether it should be added to or subtracted from the initial angle of the radius vector $\vec{R_{(t_n)}}$.
    
### <center>Step 2<center>
    
The sign of $\Delta \phi_{(t_n)}$ depends on three factors: the position of the vertical line, i.e., $x_l$, the direction of the spiral, i.e., the sign of the angular velocity $w$, and the position of the $y_{(t_n)}$-coordinate of the initial vector. Since there are two possible values for the sign of each of the three factors, we can formulate another $binary function$. Let's call it $A^{[1, -1]}_{(x_l,w,y_n)}$. Then
    
$$A^{[1, -1]}_{(x_l,w,y_{(t_n)})} = (-1)\frac{x_l}{|x_l|}\frac{w}{|w|}\frac{y_{(t_n)}}{|y_{(t_n)}|},\quad x_l \neq 0, \quad w \neq 0, \quad y_{(t_n)} \neq 0.\quad\quad\quad\quad(131)$$
    
There are $2^3$ combinations between $x_l$, $w$ and $y_{(t_n)}$. We will show only some of them.
    
$$A^{[1, -1]}_{(x_l,w,y_{(t_n)})} = -1,\quad x_l > 0,\quad w>0,\quad y_{(t_n)} > 0,\quad\quad\quad\quad(132)$$
    
$$A^{[1, -1]}_{(x_l,w,y_{(t_n)})} = 1,\quad x_l < 0,\quad w < 0,\quad y_{(t_n)} < 0,\quad\quad\quad\quad(133)$$
    
$$A^{[1, -1]}_{(x_l,w,y_{(t_n)})} = -1,\quad x_l < 0,\quad w > 0,\quad y_{(t_n)} < 0.\quad\quad\quad\quad(134)$$
    
In the above graph, all three factors are positive, $A^{[1, -1]}_{(x_l,w,y_{(t_n)})} = -1$, so the successive angles $\Delta \phi_{(t_n)}$ should be subtracted from the angle of the initial vector. In other words, the algorithm will move 'backward' along the spiral. However, if the $y_{(t_n)}$-coordinate were negative, then $A^{[1, -1]}_{(x_l,w,y_{(t_n)})} = 1$ and the algorithm would move forward along the spiral.
    
Now we add $A^{[1, -1]}_{(x_l,w,y_{(t_n)})}$ to (129).
    
$$\Delta \phi_{(t_n)} = A^{[1, -1]}_{(x_l,w,y_{(t_n)})}arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}}). \quad\quad\quad\quad(135)$$
    
We transform $\Delta \phi $ into $t$ using the relationship between the angle and the angular velocity through $t$.
    
$$\Delta \phi = wt,\quad\quad\quad\quad(136)$$
    
$$t_{(n)} = \frac{\Delta \phi_{(t_{(n-1)})}}{|w|}.\quad\quad\quad\quad(137)$$
    
In (137), we directly recorded the current time interval $t_{(n)}$, which the algorithm returns as a function of $t_{(n-1)}$ from the previous iteration. Or, as in the situation shown in the above graph, as a function of the initial time $T_n$, provided by the main numerical sequence {$T_n$}. We also take the absolute value of the angular velocity $w$, as the sign of $t_{(n)}$ will be determined by the sign of $\Delta \phi_{(t_{(n-1)})}$.
    
Substitute (135) into (137)
    
$$t_{(n)} = \frac{A^{[1, -1]}_{(x_l,w,y_{(t_{(n-1)})})}arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}})}{|w|}.\quad\quad\quad\quad(138)$$
    
To see the full form of the formula, we substitute $A^{[1, -1]}_{(x_l,w,y_{(t_{(n-1)})})}$.
    
$$t_{(n)} = \frac{(-1)\frac{x_l}{|x_l|}\frac{w}{|w|}\frac{ vt_{(n-1)}sin(\theta_0 + wt_{(n-1)})}{| vt_{(n-1)}sin(\theta_0 + wt_{(n-1)})|}arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}})}{|w|}.\quad\quad\quad\quad(139)$$
    
We are almost ready with the formula for the n-th intersection point. The second-to-last step is to add the formula for $t_n$ to the formula for the intersection points of the spiral with the ordinate axis {$T_n$}.
    
First, we need to establish the correct notation. Let's go back once more to the table showing the two-dimensional numerical sequence.

```python
show_2d_sequence_table()
```

The capital $T_n$ denotes the successive moments when the spiral vector intersects the ordinate axis. The lowercase $t_n$ indicates the step number of the spiral vector's approach to its intersection point with the line. Since the numerical sequence is two-dimensional, we will assign a double index to $t$ - $t_{nm}$. As seen in the table, the first index $n$ represents the n-th intersection point where the algorithm starts, and the second index $m$ represents the iteration number. For example, $t_{35}$ denotes the third intersection point with five iterations of the algorithm.

$$t_n = T_n,\quad\quad\quad\quad(140)$$

$$t_{nm} = T_{nm}.\quad\quad\quad\quad(141)$$

The entire algorithm consists of two parts:
1. finding the n-th intersection point $t_n$ of the spiral with the ordinate axis, and
2. adding the sum of the intervals $t_m$.

$$\{t_{nm}\} = T_n + \sum_{m=1}^{i}\{t_m\},\quad\quad\quad\quad(142)$$

$$\{t_{nm}\} = T_n + \sum_{m=1}^{i}\{\frac{(-1)\frac{x_l}{|x_l|}\frac{w}{|w|}\frac{ vt_{(m-1)}sin(\theta_0 + wt_{(m-1)})}{| vt_{(m-1)}sin(\theta_0 + wt_{(m-1)})|}arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}})}{|w|}\},\quad x_l \neq 0, \quad w \neq 0, \quad t >0,\quad i >m, \quad t_{n0} = T_n. \quad\quad\quad\quad(143)$$

Here, we changed the index $n$ of $t_n$ from (139) to $m$, introduced an upper bound $i$ on the sum $\sum_{m=1}^{i}\{t_m\}$, where $i$ is an arbitrary number greater than $m$ and represents the level of accuracy with which we choose to calculate the approximation towards the intersection point. In the code itself, it is much more convenient to define a level of accuracy - let's say to the 5th decimal place - since for different regions of the spiral, this accuracy is achieved with a varying number of iterations $i$.

The sum starts from $m=1$, but we need to define a value for $t_{n0}$ - when $m = 0$, as this is the initial value $T_n$ entered into the algorithm.

Finally, in the formula (143), we need to substitute $T_n$ from (117).

$$\{t_{nm}\} = \frac{\Delta \theta_{(w, k)} + (n-1)\pi}{|w|} +\sum_{m=1}^{i}\{\frac{(-1)\frac{x_l}{|x_l|}\frac{w}{|w|}\frac{ vt_{(m-1)}sin(\theta_0 + wt_{(m-1)})}{| vt_{(m-1)}sin(\theta_0 + wt_{(m-1)})|}arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}})}{|w|}\}.\quad\quad\quad\quad(144)$$

Substitute $\Delta \theta_{(w, k)}$ from (114)

$$\{t_{(n,k,w,v,x_l,i)_{nm}}\} = \frac{\frac{1 - \frac{w}{|w|}}{2}(\lceil k \rceil - \lfloor k \rfloor)(\pi(1 - \lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2})) +\frac{\pi}{2}(\lfloor k + 1 \rfloor- k + \frac{1 + (-1)^{\lfloor k+1\rfloor}}{2}) + (n-1)\pi}{|w|} + \\ + \sum_{m=1}^{i}\{\frac{(-1)\frac{x_l}{|x_l|}\frac{w}{|w|}\frac{ vt_{(m-1)}sin(\theta_0 + wt_{(m-1)})}{| vt_{(m-1)}sin(\theta_0 + wt_{(m-1)})|}arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}})}{|w|}\},\quad \\n \in \mathbb{N},\quad 0 <= k<4, \quad w \neq 0, \quad, v \neq 0,\quad x_l \neq 0, \quad i>m,\quad t_{n0} = T_n.\quad\quad\quad\quad(145)$$

This is the formula for the two-dimensional numerical sequence of the intersection points of a spiral and a line as a function of the initial angle $\theta$ measured in $k\frac{\pi}{2}$ radians, the angular velocity $w$, the speed $v$ of the spiral vector $\vec{R}$, the point $x_l$ where the vertical line is defined, and the number of iterations $i$, which determines the accuracy of the calculation of the intersection points.

In short, the two-dimensional numerical sequence $\{t_{(n,k,w,v,x_l,i)_{nm}}\}$ provides the $m$-th approximation of the $n$-th intersection point as a function of the specified variables.

Let’s take another look at the short version of this formula.

$$\{t_{nm}\} = T_n + \sum_{m=1}^{i}\{t_m\},\quad\quad\quad\quad(142)$$


The first term $T_n $ defines the intersection points of the spiral with the ordinate axis, or in other words, the rows $n$ of the two-dimensional numerical sequence $\{t_{nm}\} $. The second term after the summation sign $\{t_m\}$ defines the columns $m$. Тhere is one case that limits the domain of definition of $\{t_{nm}\}$, although in this case, a well-defined numerical sequence still exists. This is the case when $x_l = 0$. In this scenario, the vertical line coincides with the ordinate axis, where the intersection points are precisely defined by the first term. Let's review the formula for the second term $\{t_m\}$ from (139).

$$t_{(n)} = \frac{(-1)\frac{x_l}{|x_l|}\frac{w}{|w|}\frac{ vt_{(n-1)}sin(\theta_0 + wt_{(n-1)})}{| vt_{(n-1)}sin(\theta_0 + wt_{(n-1)})|}arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}})}{|w|}.\quad\quad\quad\quad(139)$$

(Here, we changed the index $n$ to $m$). It can be seen that $x_l$ appears once as a factor in the denominator. Hence the restriction that it cannot have a value of 0. When $x_l=0$, the second term $t_{(m)}$  must disappear, leaving only the first term, which defines the intersection points with the ordinate axis. In other words, $x_l$ has the "power" to change the mathematical structure from a two-dimensional to a one-dimensional numerical sequence. However, if the formula is left without the necessary correction and fed into a computer for calculation, the program will produce an error - division by 0. This can be avoided by creating another $binary function$. We'll call this function $X^{[1, x_l]}_{(x_l)}$, and it will be a function of $x_l$ that returns 1 if $x_l = 0$  and $x_l$ if $x_l \neq 0$.

$$X^{[1, x_l]}_{(x_l)} = \frac{1}{x_l^{0^{|x_l|} -1}},\quad\quad\quad\quad(146)$$

$$X^{[1, x_l]}_{(x_l)} = 1,\quad x_l = 0,\quad\quad\quad\quad(147)$$

$$X^{[1, x_l]}_{(x_l)} = x_l,\quad x_l \neq 0.\quad\quad\quad\quad(148)$$

In the denominator of (146), we take the absolute value of $x_l$ because zero cannot be raised to a negative power.

Now we will substitute $x_l$ in denominator in (139) with the binary function $X^{[1, x_l]}_{(x_l)}$.


$$t_{(m)} = \frac{(-1)\frac{x_l}{|X^{[1, x_l]}_{(x_l)}|}\frac{w}{|w|}\frac{vt_{(m-1)}sin(\theta_0 + wt_{(m-1)})}{|vt_{(m-1)}sin(\theta_0 + wt_{(m-1)})|}arccos(\frac{ y_{(t_n)}^2 + x_l^2 + (vt_n)^2 - (x_l - x_{(t_n)})^2}{2vt_n\sqrt {y_{(t_n)}^2 + x_l^2}})}{|w|}.\quad\quad\quad\quad(149)$$

The formula will work the same way for any $x_l \neq 0$. When $x_l = 0$

$$\frac{x_l}{|x_l|} = \frac{x_l}{|\frac{1}{x_l^{0^{|x_l|} -1}}|} = \frac{0}{1} = 0,\quad\quad\quad\quad(150)$$

and then

$$t_{(m)} = 0. \quad\quad\quad\quad(151)$$

When $t_{(m)} = 0$, only the first term with the intersection points on the ordinate remains in the formula (142).

$$\{t_{nm}\} = T_n,\quad\quad\quad\quad(152)$$

In conclusion, the binary function $X^{[1, x_l]}_{(x_l)}$ allows the domain of definition for (142) to be extended to include the case when the vertical line coincides with the ordinate axis. This enables the calculation of the intersection points without the computational machine halting due to a division by zero error.

Before deriving the general solution, we need to address a few more details. Let's take a look at how (142) appears graphically.

```python
v_line_x = -3
v = 1
w = 1
theta_0 = 3*np.pi/2
show_all_intersects(v,w, theta_0, v_line_x,
                    show_line_and_itrscts=True,
                    show_labels=True,
                    show_nm_labels=True)
```

The intersection points are shown in the form $(n:m)$. As can be seen from the non-increasing number of $m$ at each subsequent forward point $n$, the algorithm requires fewer iterations to reach the intersection point with the desired precision, which is set by default to 5 decimal places in this case. It can be concluded that for large values of $n$, the algorithm might perform just a single iteration to reach the intersection point. This makes sense because, as shown in the graph, the curve of the spiral becomes less curved as it moves away from the origin, eventually transforming into a straight line. This means that the $y$-coordinates of the intersection points on the vertical line will equal the $y$-coordinates of the same points on the ordinate axis. The spiral's curve is most pronounced near its origin. There's a certain analogy between this and the curvature of spacetime, which becomes stronger as one approaches a massive object.

The fact that the $y$-coordinates of the intersection points of the spiral with the line will become equal in value to the $y$-coordinates of the intersection points of the spiral with the ordinate axis can be easily demonstrated graphically and algebraically. Let's express this statement in mathematical form.

$$\lim_{n\to\infty} |y_{l_{(nm)}}|= |y_{o_{(n)}}|, \quad n\in \mathbb{N},\quad\quad\quad\quad(153)$$


where $y_{l_{(nm)}}$ is the $y$-coordinate of the intersection point between the spiral and the vertical line at the point $x_l$, $y_{o_{(n)}}$ is the $y$-coordinate of the intersection point of the spiral with the ordinate axis, and $n$ is the number of the corresponding intersection point. We take the absolute values of these points since we are actually comparing lengths, as shown in the graph above, where both points take negative values for even values of $n$.


The algebraic proof is as follows: when $x_l = 0$, the vertical line coincides with the ordinate axis, and therefore, the intersection points also coincide. When $x \neq 0$ we observe that the number of iterations of the algorithm decreases, i.e., the number of columns $m$ in the two-dimensional numerical sequence decreases. This directly implies that the distance between the two $y$-coordinates also decreases. On the other hand, the limit of the sequence created by the algorithm is an irrational number and can be approached infinitely by the sequence. This means that

$$|y_{l_{(nm)}}|<|y_{o_{(n)}}|, \quad \forall n, \quad x \neq 0.\quad\quad\quad\quad(154)$$

The inequality (154) is a strict inequality for two reasons. Let's look at the graph below where $x_l$ is a sufficiently large number given certain values for $v$ and $w$.



```python
v_line_x = 18
show_all_intersects(v,w, theta_0, v_line_x,
                    show_line_and_itrscts=True,
                    show_labels=True,
                    show_nm_labels=True)
```

As explained in the algorithm section, when the line does not touch any of the spiral turns, the radius vector "lands" on the $x$-axis. This means that technically, these points are not intersection points of the line and the spiral but, on the other hand, they are calculated from the algorithm itself. Clearly, at these points 

$$y_{l_{(n)}} = 0,\quad\quad\quad\quad(155)$$

therefore

$$|y_{l_{(n)}}|<|y_{o_{(n)}}|, \quad \exists n.\quad\quad\quad\quad(156)$$

This is the first reason.

The second reason is that the spiral curve increasingly straightens as it approaches infinitely close to the straight line. This straight line is actually the horizontal line at the point $y_{o_{(n)}}$. The point $y_{o_{(n)}}$ - as a function of $t$ - is the input parameter that the algorithm uses to start approaching the intersection point. As already demonstrated, the nature of the algorithm ensures that each subsequent value of $y_{l_{(nm)}}$ is smaller than the previous one. 

$$|y_{l_{(nm)}}| < |y_{l_{(nm+1)}}|.\quad\quad\quad\quad(154)$$

This is evident from the fact that, for certain values of $n$, the algorithm "lands" the intersection point on the $x$-axis, i.e., at $y_{l_{(nm)}}=0$. Therefore, the first value obtained as a result of the first iteration will always be smaller than the initial $y_{l_{n}}$. Formally

$$|y_{l_{(nm)}}| < |y_{l_{(nm+1)}}| < |y_{o_{(n)}}|,\quad\quad\quad\quad(157)$$

therefore

$$|y_{l_{(nm)}}|<|y_{o_{(n)}}|, \quad \forall n, \quad x \neq 0.\quad\quad\quad\quad(158)$$

therefore

$$\lim_{n\to\infty} |y_{l_{(nm)}}|= |y_{o_{(n)}}|, \quad n\in \mathbb{N},\quad\quad\quad\quad(153)$$

which was to be proven.
